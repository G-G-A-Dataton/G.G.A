# 01_merge_and_test_mustafa.ipynb

Bu notebook, `src/data.py` modülündeki veri yükleme ve birleştirme (merge) akışını test etmek ve doğrulamak amacıyla oluşturulmuştur.

```python
import sys
import os
import pandas as pd

# Proje kök dizinini Python yoluna ekliyoruz
sys.path.append(os.path.abspath('../'))
from src.data import load_terms, load_items, merge_pairs
```

## 1. Veri Yükleme ve Optimizasyon Testleri

```python
data_dir = '../datasets'
terms_path = os.path.join(data_dir, 'terms.csv')
items_path = os.path.join(data_dir, 'items.csv')
train_pairs_path = os.path.join(data_dir, 'training_pairs.csv')
test_pairs_path = os.path.join(data_dir, 'submission_pairs.csv')
```

```python
print("Sorgu (terms) tablosu yükleniyor...")
terms_df = load_terms(terms_path)
print(f"Terms Boyutu: {terms_df.shape}")
terms_df.info()
```

```python
print("Ürünler (items) tablosu yükleniyor (bellek dostu tiplerle)...")
items_df = load_items(items_path)
print(f"Items Boyutu: {items_df.shape}")
items_df.info()
```

## 2. Eşleşme Çiftlerini Birleştirme (Merge) Denemesi

```python
print("Eğitim çiftleri birleştiriliyor...")
train_df = merge_pairs(train_pairs_path, terms_df, items_df, is_train=True)
print(f"Train Merged Boyutu: {train_df.shape}")
train_df.head()
```

```python
print("Test çiftleri birleştiriliyor...")
test_df = merge_pairs(test_pairs_path, terms_df, items_df, is_train=False)
print(f"Test Merged Boyutu: {test_df.shape}")
test_df.head()
```

## 3. Bellek Kullanımı Doğrulaması

```python
print("Train Merged DataFrame Bellek Tüketimi:")
train_mem = train_df.memory_usage(deep=True).sum() / (1024 ** 2)
print(f"{train_mem:.2f} MB")

print("Test Merged DataFrame Bellek Tüketimi:")
test_mem = test_df.memory_usage(deep=True).sum() / (1024 ** 2)
print(f"{test_mem:.2f} MB")
```
